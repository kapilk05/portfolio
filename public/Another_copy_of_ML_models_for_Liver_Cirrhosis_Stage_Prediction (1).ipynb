```python
label_encoder = LabelEncoder()

# # Fit the encoder to your training labels and transform them
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.fit_transform(y_test)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
xgb_model = xgb.XGBClassifier(objective='multi:softprob', num_class=3, eval_metric='mlogloss')
xgb_model.fit(X_train, y_train)

# Generate class probabilities as output features for GRU
xgb_train_probs = xgb_model.predict_proba(X_train)  # Shape: (n_samples, 3)
xgb_test_probs = xgb_model.predict_proba(X_test)
```

```python
import numpy as np

# Concatenate XGBoost output with original features
X_train_combined = np.hstack((X_train, xgb_train_probs))  # Shape: (n_samples, n_features + 3)
X_test_combined = np.hstack((X_test, xgb_test_probs))
```

```python
X_train_gru = np.expand_dims(X_train_combined, axis=1)  # Shape: (n_samples, 1, n_features + 3)
X_test_gru = np.expand_dims(X_test_combined, axis=1)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define LSTM model
lstm_model = Sequential([
    LSTM(512, return_sequences=True, input_shape=(X_train_gru.shape[1], X_train_gru.shape[2])),
    Dropout(0.3),  # Optional dropout for regularization
    LSTM(256, return_sequences=True),  # Second LSTM layer
    Dropout(0.3),  # Optional dropout
    LSTM(128),  # Third LSTM layer
    Dropout(0.3),  # Optional dropout
    Dense(64, activation='relu'),  # Fully connected layer
    Dense(3, activation='softmax')  # Output layer for 3 classes
])

# Compile the LSTM model
lstm_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the LSTM model
lstm_model.fit(X_train_gru, y_train, epochs=6, batch_size=32, validation_split=0.2)

# Evaluate on test data
y_pred_lstm = np.argmax(lstm_model.predict(X_test_gru), axis=-1)
print(classification_report(y_test, y_pred_lstm))
print("LSTM Model Accuracy:", accuracy_score(y_test, y_pred_lstm))
```
